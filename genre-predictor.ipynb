```python
data = []

for album in albums:
    # make a zero array of length genres ex [0, 0, 0, 0, 0, 1]
    # with the 1 indiciating which genre it is
    genre_array = np.zeros(len(genres))
    genre_array[genres.index(album[1])] = 1

    results = sp.search(q=f'artist:{album[0][0]} album:{album[0][1]}', type='album')
    album_id = results['albums']['items'][0]['id']
    tracks = sp.album_tracks(album_id)['items']

    for track in tracks:
        track_id = track['id']
        f = sp.audio_features(track_id)[0]
        song_data = np.array([
            f['danceability'], 
            f['energy'], 
            f['loudness']/loudness_max, 
            f['speechiness'], 
            f['acousticness'], 
            f['instrumentalness'], 
            f['liveness'], 
            f['valence'], 
            f['tempo']/tempo_max
        ])
        data.append((song_data, genre_array))
```

```python
with open('data', 'rb') as f: 
    data = pickle.load(f) 

np.random.shuffle(data)
print(data[:5])

print(len(data))

loudness = [d[0][2] for d in data]
print('highest loudness', max(loudness))

tempo = [d[0][8] for d in data]
print('highest tempo', max(tempo))
```

```python
model = nn.Sequential(
  nn.Linear(9, 128),
  nn.ReLU(),
  nn.Linear(128, 128),
  nn.ReLU(),
  nn.Linear(128, 4),
  nn.Softmax(-1)
  # nn.Sigmoid()
)
optimizer = torch.optim.SGD(model.parameters(), lr=1e-1)
```

```python
for epoch in range(1, 51):
  losses = []
  for x, y in data:
    x = torch.tensor(x).reshape(9).float()
    y = torch.tensor(y).reshape(4).float()

    guess = model(x)
    # print(x, y, guess)

    loss = nn.functional.huber_loss(guess, y)

    optimizer.zero_grad()
    loss.backward()
    optimizer.step()


    losses.append(loss.item())
  if epoch % 5 == 0:
    print("LOSS", np.mean(losses))
```

```python
def get_genre(name, artist):
  results = sp.search(q="track:" + name + " artist:" + artist, type="track")
  track_id = results['tracks']['items'][0]['id']

  f = sp.audio_features(track_id)[0]
  song = np.array([
      f['danceability'], 
      f['energy'], 
      f['loudness']/loudness_max, 
      f['speechiness'], 
      f['acousticness'], 
      f['instrumentalness'], 
      f['liveness'], 
      f['valence'], 
      f['tempo']/tempo_max
  ])
  song = torch.tensor(song).reshape(9).float()

  t = model(song)
  value = (t == max(t)).nonzero(as_tuple=True)[0].detach()

  to_return = (f'{name} by {artist} is {genres[value]} ({round(int(10000*max(t)))/100}% confidence)\n')
  
  if max(t) < .6:
    t_list = list(t)
    t_list[t_list.index(max(t_list))] = torch.tensor(0)
    secondary_genre = genres[t_list.index(max(t_list))]
    to_return += f'Secondary genre: {secondary_genre} ({round(int(10000*max(t_list)))/100}% confidence)\n'

  return to_return
  
```
